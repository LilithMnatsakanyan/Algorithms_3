```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
Graph = nx.DiGraph()
Graph.add_nodes_from([1,2,3,4,5])
edges_with_weights = [(1,2,0.2),(1,3,0.5),(2,4,1),(3,5,0.8),(4,5,0.4),(3,4,0.77)]
Graph.add_weighted_edges_from(edges_with_weights)
```

```python
#Graph.add_edges_from([(1,2),(1,3),(2,4),(3,5),(4,5),(3,4)])
plt.figure(figsize=(5,5), facecolor='white')
pos = nx.spring_layout(Graph, seed=17)
nx.draw(Graph, pos, node_color='lightgreen', edge_color='gray', node_size=1000, font_size=15, with_labels=True, arrows=True)
edge_labels = {(u, v): w for u, v, w in edges_with_weights}
nx.draw_networkx_edge_labels(Graph,pos,edge_labels = edge_labels,font_color = 'lightpink',font_size = 12)
plt.savefig("images/graph_plt.png", format='png', bbox_inches='tight', facecolor='white')
plt.show()
```
